```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
from scipy import linalg as la
from numpy.linalg import matrix_power
import networkx as nx
from scipy.spatial import ConvexHull
```

```python
n = 7

capacity = 5

values = np.array([78, 63, 150, 135, 120, 40, 80])
weights = np.array([2,1,5,3,2,1,4])

x = cp.Variable(n, boolean = True)

objective = cp.Maximize(x @ values.T)

constraints = [x @ weights.T <= capacity]

prob = cp.Problem(objective, constraints)

opt_val = prob.solve(solver=cp.GUROBI)

print(f'x* = {(tuple(x.value))}')
print(f'The optimal value is {opt_val}')
```

## Q3 code

```python
def checkMatrix(A):
    if len(A.shape) != 2:
        raise Exception('Matrix is not 2 dimensional')
    if A.shape[0] != A.shape[1]:
        raise Exception('Matrix is not square')
    if A.shape[0] < 2:
        raise Exception('Matrix dimensions should be nxn, n>=2')
```

```python
def objectiveFunction(A, X, n):
    objectiveTerms = []
    for i in range(n):
        for j in range(n):
            objectiveTerms.append(A[i][j]*X[i][j])        
    objective = cp.Minimize(sum(objectiveTerms))
    return objective
```

```python
def entryConstraints(X, n):
    return [sum(X[i])==1 for i in range(n)]
```

```python
def exitConstraints(X, n):
    return [sum(X[:,j])==1 for j in range(n)]
```

```python
def solve(objective, constraints, X):
    prob = cp.Problem(objective,constraints)
    prob.solve()
    solution = X.value
    return solution
```

```python
def findCycles(X):
    Graph = nx.DiGraph(X)
    simple =  nx.simple_cycles(Graph)
    cycles = [list(cycle) for cycle in simple]
    return cycles
```

```python
def cycleLength(cycles, A):
    lengths = []
    length = 0
    for cycle in cycles:
        for idx, node in enumerate(cycle):
            if idx == len(cycle)-1:
                length+=A[node][cycle[0]]
                lengths.append(length)
                length = 0
            else:
                length+=A[node][cycle[idx+1]]
    return lengths
```

```python
def printSolution(cycles, lengths):
    for idx, cycle in enumerate(cycles):
        print(f'{idx+1}. {cycle} with a total length of {lengths[idx]}')
    if len(cycles) == 1:
        print(f'Found a feasible solution to the TSP problem with value of {lengths[0]}')
    else:
        print(f'Found an infeasible solution to the TSP problem with value of {sum(lengths)}')
```

```python
#print cycle given list of nodes example: given [0,1,2,3] print the graph 0->1->2->3->0 using nx 
def printCycle(cycle):
    G = nx.DiGraph()
    for i in range(len(cycle)):
        G.add_edge(cycle[i], cycle[(i+1)%len(cycle)])
    nx.draw(G, with_labels = True, node_color = 'lightblue', node_size = 500, font_weight = 'bold', font_color = 'black', font_size = 10, edge_color = 'black', pos = nx.circular_layout(G))
    plt.show()
```

```python
def TSP_relaxation(A):
    checkMatrix(A)  
    n = len(A)
    X = cp.Variable((n,n), boolean = True)
    objective = objectiveFunction(A, X, n)
    constraints = entryConstraints(X, n) + exitConstraints(X, n)
    solution = solve(objective, constraints, X)
    cycles = findCycles(solution)
    cycleLengths = cycleLength(cycles, A)
    return cycles, cycleLengths
```

```python
M = 1000000
A = np.array([[M, 13, 7, 5, 19],
              [14, M, 9, 3, 11],
              [7, 8, M, 2, 16],
              [4, 2, 1, M, 12],
              [10, 8, 17, 12, M]])
```

```python
Solution = TSP_relaxation(A)
printSolution(Solution[0], Solution[1])
for cycle in Solution[0]:
    printCycle(cycle)
```
