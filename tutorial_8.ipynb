```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
import os
```

```python
feasible = [(18/7,23/7), (3,0), (0,2), (0,0), (6, 1)]
hull = ConvexHull(feasible)
hull_points = np.array(feasible)[hull.vertices]
plt.fill(hull_points[:, 0], hull_points[:, 1], color='dodgerblue', alpha=0.1)
feasible.remove((6, 1))
x = np.linspace(-2000, 16, 2000)
y1 = 0.5*x+2
y2 = -0.66666*x+5 
y3 = 0.33333*x-1

# plot constraints

plt.plot(x, y1, color = 'black')
plt.plot(x, y2, color = 'black')
plt.plot(x, y3, color = 'black')
plt.axhline(y=0, color = 'black')
plt.axvline(x=0, color = 'black')
#contour lines
c1 = -x + 2.5
c2 = -x + 3.625
c3 = -x + 4.75
c4 = -x + 5.875
c5 = -x + 7

plt.plot(x, c1, color = 'gainsboro', linestyle = 'dashed')
plt.plot(x, c2, color = 'gainsboro', linestyle = 'dashed')
plt.plot(x, c3, color = 'gainsboro', linestyle = 'dashed')
plt.plot(x, c4, color = 'gainsboro', linestyle = 'dashed')
plt.plot(x, c5, color = 'gainsboro', linestyle = 'dashed')

for point in feasible:
    plt.scatter(*point, color = 'blue', zorder = 5, s = 80)

plt.scatter(6, 1, color='crimson', s=150, marker='*', zorder=2, label='Optimum')
plt.text(5.5, 1 + 0.35, f'$({6:.2f}, {1:.2f})$', color='black', zorder=2, fontsize=10)


plt.scatter(-100, 100, color = 'blue', label = 'Basic Feasible Solutions')

plt.quiver(
            0, 0,
            2, 2,
            angles='xy', scale_units='xy', scale=1.6, color='navy', label='Gradient', zorder=3
            )

plt.xlim(-0.2,8)
plt.ylim(-0.2,6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper right')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
os.makedirs('plots', exist_ok=True)
plt.savefig('plots/graph.pgf')
plt.show()
```

```python
feasible = [(18/7,23/7), (3,0), (0,2), (0,0), (63/13,24/39), (4.5, 2)]
infeasible = [(-18,-7), (4,4), (6,1), (5,0), (0,5), (0,-1), (7.5,0)]
hull = ConvexHull(feasible)
hull_points = np.array(feasible)[hull.vertices]
plt.fill(hull_points[:, 0], hull_points[:, 1], color='dodgerblue', alpha=0.1)
feasible.remove((4.5, 2))
x = np.linspace(-2000, 16, 2000)
y1 = 0.5*x+2
y2 = -0.66666*x+5 
y3 = 0.33333*x-1
y4 = -4*x+20

# plot constraints

plt.plot(x, y1, color = 'black')
plt.plot(x, y2, color = 'black')
plt.plot(x, y3, color = 'black')
plt.plot(x, y4, color = 'black')
plt.axhline(y=0, color = 'black')
plt.axvline(x=0, color = 'black')
#contour lines
c1 = -x + 2.5
c2 = -x + 3.5
c3 = -x + 4.5
c4 = -x + 5.5
c5 = -x + 6.5

plt.plot(x, c1, color = 'gainsboro', linestyle = 'dashed')
plt.plot(x, c2, color = 'gainsboro', linestyle = 'dashed')
plt.plot(x, c3, color = 'gainsboro', linestyle = 'dashed')
plt.plot(x, c4, color = 'gainsboro', linestyle = 'dashed')
plt.plot(x, c5, color = 'gainsboro', linestyle = 'dashed')

for point in feasible:
    plt.scatter(*point, color = 'blue', zorder = 5, s = 80)
for point in infeasible:
    plt.scatter(*point, color = 'lightblue', zorder = 5, s = 80)

plt.scatter(4.5, 2, color='crimson', s=150, marker='*', zorder=2, label='Optimum')
plt.text(4.5 + 0.2, 2 + 0.05, f'$({4.5:.2f}, {2:.2f})$', color='black', zorder=2, fontsize=10)


plt.scatter(-100, 100, color = 'blue', label = 'Basic Feasible Solutions')
plt.scatter(-100, 100, color = 'lightblue', label = 'Basic Infeasible Solutions')

plt.quiver(
            0, 0,
            2, 2,
            angles='xy', scale_units='xy', scale=1.6, color='navy', label='Gradient', zorder=3
            )
plt.quiver(0,0,0,1.8, color = 'darkviolet', label = 'Path Via Simplex', zorder = 3, scale = 8)
plt.quiver(0,2,81/35,8.1/7, color = 'darkviolet', zorder = 3, scale = 8)
plt.quiver(18/7,23/7,24.3/14,-8.1/7, color = 'darkviolet', zorder = 3, scale = 8)
plt.quiver(6,1,-1.28,0.85, color = 'darkorange', label = 'Path via Dual simplex', zorder = 3, scale = 8)

plt.xlim(-0.2,8)
plt.ylim(-0.2,6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper right')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
os.makedirs('plots', exist_ok=True)
plt.savefig('plots/graph.pgf')
plt.show()
```

```python
#plotting each line and highlighting the min of them
plt.figure(figsize = (10, 10))
x = np.linspace(0, 1, 1000)
y1 = 3*x - 1
y2 = -2*x + 1
y3 = -4*x + 2
y = np.minimum(y1, y2)
y = np.minimum(y, y3)
#plot the x_1 axis

#plot the scatter point of the maxima of the function
#plot the point last to make it appear on top of the other lines
plt.scatter(0.4, 0.2, color = 'red', s=200, marker='*', label = 'Optimum: $x_1=0.4, v=0.2$', zorder=3)
plt.axvline(x=0, color = 'black', linestyle='--')
plt.axvline(x=1, color = 'black', linestyle='--')
plt.plot(x, y1, label='$3x_1-1$', linestyle='--', color = 'orange')
plt.plot(x, y2, label='$-2x_1+1$', linestyle='--', color = 'forestgreen')
plt.plot(x, y3, label='$-4x_1+2$', linestyle='--', color = 'magenta')
plt.plot(x, y, label='$min\{3x_1-1, -2x_1+1, -4x_1+2\}$', color = 'Navy', linewidth=2.5)
#axis labels rotate the y label by 90 degress clockwise

plt.xlabel('$x_1$', fontsize=20)
plt.ylabel('$v$', fontsize=20, rotation=0)
plt.legend(loc='upper right', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig('plots/min_function.pgf')
plt.show()
```
